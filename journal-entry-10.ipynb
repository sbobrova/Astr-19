```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.random.normal(loc=0,scale=1,size=1000)
```

```python
plt.figure(figsize=(6,4))
plt.hist(data, bins=100, color='r', edgecolor='b')

plt.xlabel("random value")
plt.ylabel("frequency")
plt.title("normally distributed random numbers histogram")

plt.savefig("normally_distributed_histogram.pdf")

plt.show()
```
